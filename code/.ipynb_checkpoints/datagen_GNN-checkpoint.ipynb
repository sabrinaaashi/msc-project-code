### Test using mlp to fit a cubic

```python
x = torch.rand(100,1)
y = -x**3 + x**2 - x
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

class SimpleMLP(nn.Module):
    def __init__(self):
        super(SimpleMLP, self).__init__()
        self.mlp = nn.Sequential(
            nn.Linear(1, 4),
            nn.ReLU(),
            nn.Linear(4, 8),
            nn.ReLU(),
            nn.Linear(8, 4),
            nn.ReLU(),
            nn.Linear(4, 1)
        )

    def forward(self, x):
        return self.mlp(x)

model = SimpleMLP()

# Set hyperparameters
learning_rate = 0.01
num_epochs = 1000

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Split the dataset into training and validation sets
train_size = int(0.8 * len(x))
val_size = len(x) - train_size
x_train, x_val = torch.split(x, [train_size, val_size])
y_train, y_val = torch.split(y, [train_size, val_size])

# Training loop with early stopping
num_epochs = 5000
patience = 10
best_val_loss = float('inf')
losses = []
early_stopping_counter = 0

for epoch in range(num_epochs):
    # Forward pass
    outputs = model(x_train)
    loss = criterion(outputs, y_train)
    
    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # Store loss for visualization
    losses.append(loss.item())
    
    # Validation
    model.eval()
    with torch.no_grad():
        val_outputs = model(x_val)
        val_loss = criterion(val_outputs, y_val)
    model.train()
    
    # Print loss every 500 epochs
    if (epoch+1) % 500 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

    # Early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        early_stopping_counter = 0
        # Save the best model
        #torch.save(model.state_dict(), 'best_model.pth')
    else:
        early_stopping_counter += 1
        if early_stopping_counter >= patience:
            print(f"Early stopping,epoch {epoch}")
            break

        
# Plot training loss
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```

```python
import torch
from torch import nn
from torch.nn import Sequential, Linear, ReLU
from torch_geometric.utils import scatter
from torch_geometric.nn import MetaLayer
```

```python
class EdgeModel(torch.nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, src, dst, edge_attr, u, batch):
        # src, dst: [E, F_x], where E is the number of edges.
        # edge_attr: [E, F_e]
        # u: [B, F_u], where B is the number of graphs.
        # batch: [E] with max entry B - 1.

        return src[:, 0] * edge_attr

class NodeModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        
        num_node_features = 2
        in_channels = 3+1 
        hiddens = 2*in_channels
        #self.node_mlp = Sequential(
            #nn.Linear(in_channels, hiddens),
            #nn.ReLU(),
            #nn.Linear(hiddens, in_channels),
            #nn.ReLU(),
            #nn.Linear(in_channels, num_node_features))
        self.node_mlp = Sequential(
            nn.Linear(in_channels, hiddens),
            nn.ReLU(),
            nn.Linear(hiddens, 2*hiddens),
            nn.ReLU(),
            nn.Linear(2*hiddens, hiddens),
            nn.ReLU(),
            nn.Linear(hiddens, in_channels),
            nn.ReLU(),
            nn.Linear(in_channels, num_node_features))
        
    def forward(self, x, edge_index, edge_attr, u, batch):
        # x: [N, F_x], where N is the number of nodes.
        # edge_index: [2, E] with max entry N - 1.
        # edge_attr: [E, F_e]
        # u: [B, F_u]
        # batch: [N] with max entry B - 1.
        row, col = edge_index
        out = scatter(edge_attr, col, dim=0, dim_size=x.size(0),
                      reduce='sum')
        out = torch.cat([x, out.unsqueeze(1)], dim=1)
        return self.node_mlp(out)
```

```python
class EdgeModel(torch.nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, src, dst, edge_attr, u, batch):
        # src, dst: [E, F_x], where E is the number of edges.
        # edge_attr: [E, F_e]
        # u: [B, F_u], where B is the number of graphs.
        # batch: [E] with max entry B - 1.

        return src[:, 0] * edge_attr

class NodeModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        
        num_node_features = 2
        in_channels = 3+1 
        hiddens = 2*in_channels
        self.node_mlp = Sequential(
            nn.Linear(in_channels, hiddens),
            nn.ReLU(),
            nn.Linear(hiddens, in_channels),
            nn.ReLU(),
            nn.Linear(in_channels, num_node_features))

        
    def forward(self, x, edge_index, edge_attr, u, batch):
        # x: [N, F_x], where N is the number of nodes.
        # edge_index: [2, E] with max entry N - 1.
        # edge_attr: [E, F_e]
        # u: [B, F_u]
        # batch: [N] with max entry B - 1.
        row, col = edge_index
        out = scatter(edge_attr, col, dim=0, dim_size=x.size(0),
                      reduce='sum')
        out = torch.cat([x, out.unsqueeze(1)], dim=1)
        return self.node_mlp(out)
```

```python
def data_gen(batch_size, sim_length,**args):
    args['tsteps'] = sim_length
    N = args['N']
    dt = args['dt']
    args['T'] = int(sim_length*dt)
    # note that time steps = T/dt, dt=0.1
    control_scales = [0, 0.1, 0.01]
    while True:
        data = torch.zeros((batch_size, sim_length, 2, N))
        controls = torch.zeros((batch_size, sim_length, N))

        for b in range(batch_size):
            init = functions.set_random_initial_conds_for_nodes(bifurcation,noise_strength,sim_length,d,dt,parameters,realizations,noise)# (2,N)
            args['node_ic'] = init
            control_scale = random.choice(control_scales)
            randcontrol = torch.rand((sim_length,N))*control_scale
            controls[b, :, :] = randcontrol
            data[b, :, :, :] = functions.plot_runge_kutta(functions.ODE_FHN_network,randcontrol, **args) # gives time seris of legth sim_length

        # To return all the data at once
        # yield(data)

        # To return the data in chunks of length L
        L = 10
        num_chunks = sim_length // L
        for i in range(num_chunks):
            yield(data[:, i : i+L, :, :], controls[:, i : i+L, :])
```

```python
adj = torch.ones(n_nodes,n_nodes)
torch.diagonal(adj, 0).zero_()
edge_index = adj.nonzero().t().contiguous()
edge_attr = A[edge_index[0], edge_index[1]]
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = MetaLayer(EdgeModel(), NodeModel()).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=5e-4)
loss_fn = nn.MSELoss()
look_ahead = 10
train_losses = []

generator = data_gen(1,300,**args)
loss_value = float('inf')
max_epochs = 10000
epoch = 0
# Training loop
while loss_value > 1e-4 and epoch < max_epochs:
    # Fetch the next batch of data
    states,controls = next(generator) #[batch_size,look_ahead,2,N]
    states,controls = states.to(device),controls.to(device)
    if states.shape[0] == 1:
        states = states.squeeze(0)
        controls = controls.squeeze(0).unsqueeze(1)
    
    model.train()
    optimizer.zero_grad()
    loss = 0
    current_state = torch.cat((states[0, :], controls[0, :]), dim=0)  
    for k in range(look_ahead-1):
        prediction = model(current_state.T, edge_index, edge_attr)[0].T
        loss = loss + loss_fn(prediction, states[k + 1, :])
        current_state = torch.cat((prediction, controls[k + 1, :]), dim=0)
    loss.backward()
    optimizer.step()
    
    # Store and print the loss
    loss_value = loss.item()
    train_losses.append(loss_value)
    epoch +=1
    if (epoch + 1) % 1000 == 0: 
        print(f"Epoch {epoch + 1},Current Loss: {loss_value}")
```

```python
class EdgeModel(torch.nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, src, dst, edge_attr, u, batch):

        return src[:, 0] * edge_attr

        
class NodeModel(torch.nn.Module):
    def __init__(self, hidden_sizes):
        super().__init__()
        
        num_node_features = 2
        in_channels = 2+1 
        hiddens = 2*in_channels

        layers = []
        input_size = in_channels
        for hidden_size in hidden_sizes:
            layers.append(nn.Linear(input_size, hidden_size))
            layers.append(nn.ReLU())
            input_size = hidden_size

        layers.append(nn.Linear(input_size, num_node_features))
        
        self.node_mlp = nn.Sequential(*layers)
        
    def forward(self, x, edge_index, edge_attr, u, batch):
        row, col = edge_index
        out = scatter(edge_attr, col, dim=0, dim_size=x.size(0),
                      reduce='sum')
        out = torch.cat([x, out.unsqueeze(1)], dim=1)
        return self.node_mlp(out)
```

```python
import os
import pandas as pd

# Ensure that the 'grid_search' directory exists
if not os.path.exists('grid_search'):
    os.makedirs('grid_search')

# Define your grid
node_hidden_layer_combinations = [[6,3],[6,12,6], [6,12,24,12,6], [16,32,16], [16, 32, 64,32,16]]
learning_rates = [0.001, 0.01, 0.1]

grid = [(n_sizes, lr) for n_sizes in node_hidden_layer_combinations for lr in learning_rates]

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
optimizer_class = torch.optim.Adam
loss_fn = nn.MSELoss()
look_ahead = 10

max_epochs = 100000

train_generator = data_nocontrol(1, 400, **args)  # Assuming you've defined this already
valid_generator = data_nocontrol(1, 400, **args)  # A separate generator for validation

# Start the grid search
results = []
for node_sizes, learning_rate in grid:
    print(f"Training for Node Hidden Sizes: {node_sizes} and Learning Rate: {learning_rate}")

    # Create your models with the current hyperparameters
    edge_model = EdgeModel().to(device)
    node_model = NodeModel(node_sizes).to(device)
    model = MetaLayer(edge_model, node_model).to(device)

    optimizer = optimizer_class(model.parameters(), lr=learning_rate, weight_decay=5e-4)
    
    train_losses = []
    valid_losses = []
    epoch = 0
    loss_value = float('inf')
    
    # Your training loop remains largely unchanged
    while loss_value > 1e-6 and epoch < max_epochs:
        # Training step
        states = next(train_generator)
        states = states.to(device)
        if states.shape[0] == 1:
            states = states.squeeze(0)
        
        model.train()
        optimizer.zero_grad()
        loss = 0
        current_state = states[0, :]
        for k in range(look_ahead-1):
            prediction = model(current_state.T, edge_index, edge_attr)[0].T
            loss = loss + loss_fn(prediction, states[k + 1, :])
            current_state = prediction
        
        loss.backward()
        optimizer.step()
        
        # Store and print the loss
        loss_value = loss.item()
        train_losses.append(loss_value)
        
        # Validation step
        with torch.no_grad():
            model.eval()  # Switch the model to evaluation mode
            states = next(valid_generator)
            states = states.to(device)
            if states.shape[0] == 1:
                states = states.squeeze(0)
            
            valid_loss = 0
            current_state = states[0, :]
            for k in range(look_ahead-1):
                prediction = model(current_state.T, edge_index, edge_attr)[0].T
                valid_loss += loss_fn(prediction, states[k + 1, :])
                current_state = prediction
                
            valid_losses.append(valid_loss.item())

        epoch += 1
            
    best_train_loss = min(train_losses)
    avg_last_N_train_losses = sum(train_losses[-10000:]) / 10000
    best_valid_loss = min(valid_losses)
    avg_last_N_valid_losses = sum(valid_losses[-10000:]) / 10000
    results.append((node_sizes, learning_rate, best_train_loss, avg_last_N_train_losses, best_valid_loss, avg_last_N_valid_losses,train_losses,valid_losses))
    print(f"epoch {epoch},node sizes{node_sizes},learning rate{learning_rate},best train/val loss{best_train_loss}/{best_valid_loss},avg loss{avg_last_N_train_losses}/{avg_last_N_valid_losses}")
    
    # Save model, training losses, and validation losses if needed
    model_path = os.path.join('grid_search', f"model_node_{'_'.join(map(str, node_sizes))}_lr_{learning_rate}.pth")
    train_losses_path = os.path.join('grid_search', f"train_losses_node_{'_'.join(map(str, node_sizes))}_lr_{learning_rate}.txt")
    valid_losses_path = os.path.join('grid_search', f"valid_losses_node_{'_'.join(map(str, node_sizes))}_lr_{learning_rate}.txt")
    
    torch.save(model.state_dict(), model_path)
    


    df = pd.DataFrame({
        "train_loss": train_losses,
        "valid_loss": valid_losses
    })
    file_name = f"losses_node_{'_'.join(map(str, node_sizes))}_lr_{learning_rate}.csv"
    df.to_csv(file_name, index=False)        
```

```python
import os
import pandas as pd

# Ensure that the 'grid_search' directory exists
if not os.path.exists('grid_search'):
    os.makedirs('grid_search')

# Define your grid
node_hidden_layer_combinations = [[6,3],[6,12,6], [6,12,24,12,6], [16,32,16], [16, 32, 64,32,16]]
learning_rates = [0.001, 0.01, 0.1]
look_ahead_values = [5, 20,30,40,50]  

grid = [(n_sizes, lr, la) for n_sizes in node_hidden_layer_combinations for lr in learning_rates for la in look_ahead_values]

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
optimizer_class = torch.optim.Adam
loss_fn = nn.MSELoss()
val_look_ahead = 50

max_epochs = 100000

train_generator = data_nocontrol(1, 400, **args)  # Assuming you've defined this already
valid_generator = data_nocontrol(1, 400, **args)  # A separate generator for validation

# Start the grid search
results_2 = []
for node_sizes, learning_rate, look_ahead in grid:
    print(f"Training for Node Hidden Sizes: {node_sizes}, Learning Rate: {learning_rate}, Look Ahead: {look_ahead}")

    # Create your models with the current hyperparameters
    edge_model = EdgeModel().to(device)
    node_model = NodeModel(node_sizes).to(device)
    model = MetaLayer(edge_model, node_model).to(device)

    optimizer = optimizer_class(model.parameters(), lr=learning_rate, weight_decay=5e-4)
    
    train_losses = []
    valid_losses = []
    avg_valid_losses = []
    valid_losses_10 = []
    epoch = 0
    loss_value = float('inf')
    
    # Your training loop remains largely unchanged
    while loss_value > 1e-6 and epoch < max_epochs:
        # Training step
        states = next(train_generator)
        states = states.to(device)
        if states.shape[0] == 1:
            states = states.squeeze(0)
        
        model.train()
        optimizer.zero_grad()
        loss = 0
        current_state = states[0, :]
        for k in range(look_ahead-1):
            prediction = model(current_state.T, edge_index, edge_attr)[0].T
            loss = loss + loss_fn(prediction, states[k + 1, :])
            current_state = prediction
        
        loss.backward()
        optimizer.step()
        
        # Store and print the loss
        loss_value = loss.item()
        train_losses.append(loss_value)
        
        # Validation step
        with torch.no_grad():
            model.eval()  # Switch the model to evaluation mode
            states = next(valid_generator)
            states = states.to(device)
            if states.shape[0] == 1:
                states = states.squeeze(0)
        
            current_state = states[0, :]
            accumulated_losses = [0] * (val_look_ahead - 1)  # List to hold accumulated loss at each step
        
            for k in range(val_look_ahead-1):
                prediction = model(current_state.T, edge_index, edge_attr)[0].T
                step_loss = loss_fn(prediction, states[k + 1, :])
                accumulated_losses[k] = step_loss.item() + (accumulated_losses[k-1] if k > 0 else 0)  # accumulate losses for steps 1, 2, ..., k
                current_state = prediction
        
            avg_losses_per_step = [accumulated_losses[i] / (i+1) for i in range(val_look_ahead-1)]  # Take the average of accumulated losses
            avg_valid_losses.append(avg_losses_per_step.item())                
            valid_losses.append(accumulated_losses.item())
            valid_losses_10.append(accumulated_losses[9].item())

        epoch += 1
            
    best_train_loss = min(train_losses)
    avg_last_N_train_losses = sum(train_losses[-10000:]) / 10000
    best_valid_loss = min(valid_losses_10)
    avg_last_N_valid_losses = sum(valid_losses_10[-10000:]) / 10000
    acc_best_valid_loss = min(acc_losses[look_ahead-1] for acc_losses in valid_losses)
    acc_avg_last_N_valid_losses = sum(acc_losses[-1] for acc_losses in valid_losses[-10000:]) / 10000
    results_2.append((node_sizes, learning_rate, best_train_loss, avg_last_N_train_losses, best_valid_loss, avg_last_N_valid_losses,acc_best_valid_loss,acc_avg_last_N_valid_losses,train_losses,valid_losses))
    print(f"epoch {epoch},node sizes{node_sizes},learning rate{learning_rate},best train/val loss{best_train_loss:.5f}/{best_valid_loss:.5f},avg loss{avg_last_N_train_losses:.5f}/{avg_last_N_valid_losses:.5f},acc 500 val {acc_best_valid_loss:.5f}/{acc_avg_last_N_valid_losses:.5f}")
    
    # Save model, training losses, and validation losses if needed
    model_path = os.path.join('grid_search', f"model_node_{'_'.join(map(str, node_sizes))}_lr_{learning_rate}_la_{look_ahead}.pth")
    train_losses_path = os.path.join('grid_search', f"train_losses_node_{'_'.join(map(str, node_sizes))}_lr_{learning_rate}_la_{look_ahead}.csv")
    valid_losses_path = os.path.join('grid_search', f"valid_losses_node_{'_'.join(map(str, node_sizes))}_lr_{learning_rate}_la_{look_ahead}.csv")

    torch.save(model.state_dict(), model_path)
    
    df_train = pd.DataFrame({
        "train_loss": train_losses
    })
    df_valid = pd.DataFrame(valid_losses, columns=[f"valid_loss_step_{i+1}" for i in range(val_look_ahead-1)])
    df = pd.concat([df_train, df_valid], axis=1)
    file_name = os.path.join('grid_search', f"losses_node_{'_'.join(map(str, node_sizes))}_lr_{learning_rate}_la_{look_ahead}.csv")
    df.to_csv(file_name, index=False)        
```
